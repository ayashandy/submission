## Visualization & Explanatory Analysis

###  Bagaimana pengaruh musim terhadap jumlah penyewaan sepeda?

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=day_df['season'], y=day_df['cnt'])
plt.title('Rata-rata Penyewaan Sepeda per Musim(Season)')
plt.xlabel('Musim(Season)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.xticks([0, 1, 2, 3], ['Winter', 'Spring', 'Summer', 'Fall'])
plt.show()
```

### Bagaimana pengaruh cuaca terhadap jumlah penyewaan sepeda?

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=day_df['weathersit'], y=day_df['cnt'])
plt.title('Rata-rata Penyewaan Sepeda per Cuaca(Weather)')
plt.xlabel('Cuaca(Weather)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.xticks([0, 1, 2], ['Clear or partly cloudy', 'Cloudy', 'Light rain + Light snow'])
plt.show()
```

**Insight:**
- Dari hasil visualisasi data, dapat dilihat lebih jelas mana yang lebih banyak jumlah penyewaan sepeda.

- Pada pertanyaan 1, dapat disimpulkan bahwa musim sangat berpengaruh terhadap penyewaan sepeda, dapat dilihat pada musim panas atau summer merupakan musim paling banyak peminat penyewaan sepada, dan untuk yang paling rendah yaitu pada saat musim dingin atau winter karena kenyamanan saat bersepeda terjadi di musim yang bagus.
